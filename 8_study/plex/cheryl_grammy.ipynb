```python
import os
import shutil

def copy_files_from_m3u(m3u_file, dest_dir):
    """
    Reads an M3U file and copies each listed file to the specified destination directory.

    Args:
        m3u_file (str): Path to the M3U file.
        dest_dir (str): Path to the destination directory.
    """

    try:
        os.makedirs(dest_dir, exist_ok=True)  # Create the destination directory if it doesn't exist

        with open(m3u_file, 'r') as f:
            for line in f:
                line = line.strip()
                if line.startswith('#'):  # Skip comments
                    continue
                if os.path.isfile(line):
                    shutil.copy2(line, dest_dir)
                    print(f"Copied {line} to {dest_dir}")
                else:
                    print(f"Warning: File not found: {line}")

    except FileNotFoundError:
        print(f"Error: M3U file not found: {m3u_file}")
    except Exception as e:
        print(f"An error occurred: {e}")

# Example usage:
m3u_file = '/tmp/grammy.m3u'
dest_dir = '/tmp/cheryl_grammy'
copy_files_from_m3u(m3u_file, dest_dir)
```
