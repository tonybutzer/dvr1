```python
import time
from subprocess import Popen
import pandas as pd
```

```python
from util import run_command, read_json_to_dict
```

```python
def get_tuner_signal_strength(hdhr):
    command2 = f'hdhomerun_config {hdhr} get /tuner1/status'
    output2, returncode2 = run_command(command2)
    print(output2)
    return(output2)
```

```python
hdhr='10B16BC7'
hdip = '192.168.0.151'
get_tuner_signal_strength(hdhr)
```

```python
def get_channel_status(hdip, hdhr, chan):
    command = f'curl -so a.ts {hdip}:5004/tuner1/v{chan}?duration=3 &'
    cmd_list = command.split(' ')
    p = Popen(cmd_list) # something long running
    time.sleep(2)
    status = get_tuner_signal_strength(hdhr)
    time.sleep(1)
    # ... do other stuff while subprocess is running
    p.terminate()
    return(status)
```

```python
strength = get_channel_status(hdip, hdhr, '13.1')
```

```python
# Example usage:
file_path = 'lineup.json'  # Replace with the actual path to your JSON file
data_dict = read_json_to_dict(file_path)

#print(data_dict) 
```

```python
df = pd.DataFrame.from_dict(data_dict)
```

```python
strength_list = []
for index, row in df.iterrows():
    chan = row['GuideNumber']
    #print(chan)
    strength = get_channel_status(hdip, hdhr, chan)
    strength_list.append(f'chan={chan} {strength}')
```

```python
import pandas as pd

def create_dataframe(data):
  """Converts a list of strings into a pandas dataframe.

  Args:
    data: A list of strings, where each string represents a line of data with multiple key-value pairs separated by spaces.

  Returns:
    A pandas dataframe where each row represents a line and each column represents a key-value pair.
    The column names are the keys and the values are the corresponding values.
  """

  def parse_line(line):
    """Parses a single line of data into a dictionary of key-value pairs."""
    sdict={}
    pairs = line.strip().split()
    print(pairs)
    for pair in pairs:
        ky,val = pair.split('=')
        sdict[ky]=val
    return(sdict)
    #return {key: value for key, value in zip(pairs[::2], pairs[1::2])}

  # Parse each line into a dictionary
  data_dicts = [parse_line(line) for line in data]

  # Create a dataframe from the list of dictionaries
  df = pd.DataFrame(data_dicts)

  return df



# Create the dataframe
df = create_dataframe(strength_list)
```
